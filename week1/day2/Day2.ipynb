ZIP FUNCTION()

```python
#zip function, it combines mupltiple collection into pairs
#it groups element based on the index positions
```

```python
coders =["Kusum", "Nike", "Hari", "Belly", "Suyog"]
marks =[90, 40,50, 88,95]

list(zip(coders, marks))
```

```python
#Indexwise mapping
#index 0 has(Kusum + 90) values
#zip returns a zip object, not a list
```

```python
zip(coders, marks)
```

```python
#Zip stops at the shortest list
#if no pair then stops the short list it as the num has 2 list
alphabet = ["A", "B", "C"]
num = [80,90]
list(zip(alphabet, num))
```
